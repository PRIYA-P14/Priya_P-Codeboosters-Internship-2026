# **Practice Questions:**


How many students are from the electronics department?

What is the avg attendance of the electronics department?

Who is the student with the lowest programming score?

How many student scored above 80 in math?

What is the average total_score for students with attendance above 90%?

```python
#How many students are from the electronics department?
electronics_student=df[df['department']=='Electronics']
len(electronics_student)
```

```python
#What is the avg attendance of the electronics department?
avg_attendance=electronics_student['attendance_percentage'].mean()
print(avg_attendance)
```

```python
#Who is the student with the lowest programming score?
lowest_programming=df.loc[df['programming_score'].idxmin()]
print(lowest_programming)
```

```python
#How many student scored above 80 in math?
above_80=df[df['math_score']>80]
len(above_80)
```

```python
#What is the average total_score for students with attendance above 90%?
above_90=df[df['attendance_percentage']>90]
avg_total_score=above_90['total_score'].mean()
print(avg_total_score)
```
